Query: Which bowler bowled maximum number of dot balls

```python
import pandas as pd
```

```python
deliv_df=pd.read_csv("deliveries.csv")
```

```python
df=deliv_df.query("wide_runs>0")
```

```python
df
```

```python
bowler=df.groupby('bowler')
```

```python
bowler.aggregate('count')
```
