```python
import pandas as pd
```

```python
matches_df=pd.read_csv("matches.csv")
```

```python
df1=matches_df[matches_df['toss_winner']==matches_df['winner']]
```

```python
df1=df1.query("toss_decision=='field'")
```

```python
df2=matches_df[matches_df['toss_winner']!=matches_df['winner']]
```

```python
df2=df2.query("toss_decision=='bat'")
```

```python
df=pd.concat([df1,df2])
```

```python
players=df.groupby('player_of_match')
```

```python
players.aggregate('count')
```
