Which bowler has taken most wickets?

```python
import pandas as pd
```

```python
deliv_df=pd.read_csv("deliveries.csv")
```

```python
df=deliv_df.dropna()
```

```python
df
```

```python
bowler=df.groupby("bowler")
```

```python
bowler.aggregate('count')
```
